```python
# Importing all the python liberaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
athletes = pd.read_csv('/content/drive/MyDrive/PythonProjectsSimplilearn/OlympicsDatasetPython/athlete_events.csv')
regions = pd.read_csv('/content/drive/MyDrive/PythonProjectsSimplilearn/OlympicsDatasetPython/noc_regions.csv')
```

```python
# Few rows of dataset
athletes.head()
```

```python
# Few rows of another dataset
regions.head()
```

```python
# Joining the 2 datasets using pandas merge function(horizontal join)
athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
# Shape of dataframe
athletes_df.shape
```

```python
# Making column names consistent using rename
athletes_df.rename(columns = {'region':'Region','notes':'Notes'}, inplace = True)
athletes_df.head()
```

```python
# Using info method to see datatypes
athletes_df.info()
```

```python
# Using describe function to display statistical summary
athletes_df.describe()
```

```python
# Checking null values in dataset
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# Number of null values
athletes_df.isnull().sum()
```

```python
# Top 10 countries participating in 1896
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for top 10 countries
plt.figure(figsize=(12,6))
#plt.xticks(notation=20)
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2')
```

```python
# Age Distribution of athletes
plt.figure(figsize=(12,6))
plt.title('Age Distribution of thr athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins=np.arange(10,80,2), color='green',edgecolor='white')
plt.show()
```

```python
# Winter olympic sports
winter_sports=athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
# Summer olympic sports
summer_sports=athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
# Male and Female participants
gender_counts=athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=150, shadow=False)
plt.show()
```

```python
# Total medals
athletes_df.Medal.value_counts()
```

```python
# Total number of female athletes in each olympics
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season=='Summer')]
womenOlympics.head()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenOlympics,palette="Spectral")
plt.title('Women Participation')
plt.show()
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
plt.show()
```

```python
# Gold medal athletes
goldMedals = athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head()
```

```python
# Take only values different from NaN
goldMedals=goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Gold beyond 60
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event=goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
# Plot for sporting_event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for Athletes over 60 years age')
plt.show()
```

```python
# Gold medals from each country
goldMedals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGoldMedals=goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g=sns.catplot(x="index",y="Medal",data=totalGoldMedals,height=5,kind="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
plt.show()
```

```python
# Rio Olympics
max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year==max_year) & (athletes_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y= team_names.value_counts().head(20).index)

plt.ylabel(None)
plt.xlabel('Countrywise Medals for the year 2016')
```
